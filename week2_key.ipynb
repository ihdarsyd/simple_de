```python
# Import library
import pandas as pd

# Read Data
sales_data = pd.read_csv('./week_4/sales_data.csv')

# Get top 5 data
sales_data.head()
```

```python
def columns_requirements():
    columns = {'order_id' : int,
               'product' : object,
               'quantity_orderd' : int,
               'price_each' : float,
               'order_date' : 'datetime64[ns]',
               'street' : object,
               'city' : object,
               'zip_code' : object
               }
    return columns
```

```python
def validation_sales(sales):
    # data shape
    print(f"Row Data: {sales.shape[0]}")
    print(f"Column  Data: {sales.shape[1]}")

    # validation column
    actual_column = sales.columns
    expected_column = columns_requirements().keys()

    list_column = []
    list_not_column = []
    for column in actual_column:
        if column in expected_column:
            list_column.append(column)
        else:
            list_not_column.append(column)
    
    print(f"Sesuai dengan req: {list_column}")
    print(f"Tidak sesuai req: {list_not_column}")

    list_not_in_actual = []
    for column in expected_column:
        if column not in actual_column:
            list_not_in_actual.append(column)
    
    print(f"Kolom yang belum ada di actual_column: {list_not_in_actual}")


    # Validation data types
    actual_dtype = sales.columns
    expected_dtype = columns_requirements() 

    list_dtype_not = []
    for key, value in expected_dtype.items():
        if key in sales.columns:
            if value == sales[f'{key}'].dtype:
                print(f"{key} sudah sesuai, {value}")
            else:
                list_dtype_not.append(key)
                print(f"{key} tidak sesuai,sekarang {sales[f'{key}'].dtype} harusnya {value}")
    print("")
    print(f"Column Data Type yang tidak sesuai {list_dtype_not}")


    # Missing Value untuk setiap kolom
    percentage_missing =  {} # column:mising_value%

    for column in sales.columns:
        percentage = sales[column].isna().sum() / len(sales) * 100
        percentage_missing[column] = percentage
    
    print("")
    print(percentage_missing)


    #Data Duplicate
    print(f"Data Duplicate: {sales.duplicated(keep= False).sum() / len(sales) * 100}")
    print(f"Jumlah Data Duplicate: {sales.duplicated(keep= False).sum()}")

    
```

```python
validation_sales(sales_data)
```

```python
def handling_sales(sales):
    # pembuatan kolom baru ['street', 'city', 'zip_code'] didapatkan dari 'purchase_address'
    sales['street'] = sales['purchase_address'].str.extract(r'(\d+\s[A-Za-z0-9\s]+),')

    sales['city']= sales_data['purchase_address'].str.extract(r'(\d+\s[A-Za-z\s]+),')

    sales['zip_code'] = sales['purchase_address'].str.extract(r'([A-Z]{2}\s\d+)')


    # hapus kolom yang tidak sesuai req
    sales = sales.drop('purchase_address',axis = 1)

    # handling missing value
    sales = sales.dropna()

    # Handling data duplikat
    sales = sales.drop_duplicates(keep = 'first')

    # casting Data type ['order_id', 'quantity_orderd', 'order_date']
    sales['order_id'] = sales['order_id'].astype(int)
    sales['quantity_orderd'] = sales['quantity_orderd'].astype(int)

    sales['order_date'] = pd.to_datetime(sales['order_date'])

    return sales
```

```python
sales_transformation = handling_sales(sales_data)
```

```python
validation_sales(sales_transformation)
```

```python
# Import library
import pandas as pd

# Read Data
bank_churners = pd.read_csv('./week_4/BankChurners.csv')

# Get top 5 data
bank_churners.head()
```

```python
def columns_requirements():
    columns = {'CLIENTNUM' : int,
               'Attrition_Flag' : object,
               'Customer_Age' : int,
               'Gender' : object,
               'Dependent_count' : int,
               'Education_Level' : object,
               'Marital_Status' : object,
               'Income_Category_Min' : int,
               'Income_Category_Max' : int,
               'Card_Category' : object,
               'Months_on_book' : int,
               'Total_Relationship_Count' : int,
               'Months_Inactive_12_mon' : int,
               'Contacts_Count_12_mon' : int,
               'Credit_Limit' : float,
               'Total_Revolving_Bal' : int,
               'Avg_Open_To_Buy' : float,
               'Total_Amt_Chng_Q4_Q1' : float,
               'Total_Trans_Amt' : int,
               'Total_Trans_Ct' : int,
               'Total_Ct_Chng_Q4_Q1' : float,
               'Avg_Utilization_Ratio' : float}
    return columns
```

```python
def values_requirements():
    values = {
        'Attrition_Flag' : ['Existing Customer', 'Attrited Customer'],
        'Gender' : ['M', 'F'],
        'Education_Level' : ['High School', 'Graduate', 'Uneducated', 'Unknown', 'College', 'Post-Graduate', 'Doctorate'],
        'Marital_Status' : ['Married', 'Single', 'Divorced'],
        'Card_Category' : ['Blue', 'Gold', 'Silver', 'Platinum']
    }
    return values
```

```python
def validation_churn(data_churn):
    # Validasi Data shape
    print(f"Row data: {data_churn.shape[0]}")
    print(f"Column: {data_churn.shape[1]}")

    # validasi column
    print("")
    actual_column = data_churn.columns
    expected_column = columns_requirements().keys()
    # print(actual_column, expected_column)

    if len(actual_column) != len(expected_column):
        print("Jumlah Kolom tidak sama")
    else:
        print("Jumlah kolom sama")
    
    list_column = []
    list_not_column = []
    for column in actual_column:
        if column in expected_column:
            list_column.append(column)
        else:
            list_not_column.append(column)
    
    list_must_have = []
    for column in expected_column:
        if column not in actual_column:
            list_must_have.append(column)

    print(f"List kolom sesuai dengan expected column {len(list_column)}: {list_column}")
    print(f"List kolom tidak ada di expected column {len(list_not_column)}: {list_not_column}")
    print(f"List Kolom yang seharusnya ada {len(list_must_have)}: {list_must_have}")

    # validasi data type
    print("")
    expected_dtype = columns_requirements()
    list_not_dtype = []
    for key, value in expected_dtype.items():
        if key in actual_column:
            if data_churn[f"{key}"].dtypes == value :
                print(f"{key} sudah sesuai {value}")
            else:
                print(f"{key} tidak sesuai, sekarang {data_churn[f'{key}'].dtypes} seharusnya {value}")
                list_not_dtype.append(key)

    print(f"List kolom dengan data tipe yang tidak sesuai {len(list_not_dtype)} {list_not_dtype}")

    # validasi missing value
    print("")
    percentange_missing = {}
    for column in data_churn.columns:
        percentage = data_churn[column].isna().sum() / len(data_churn) * 100
        if percentage != 0:
            percentange_missing[column] = percentage
    print(percentange_missing)


    # validasi data duplicate
    print("")
    print(f"Data Duplikat: {data_churn.duplicated(keep = False).sum()}")

    # validasi inconsistent data
    print("")
    value_consistent = values_requirements()
    for column in value_consistent.keys():
        data_value = data_churn[column].unique()
        miss_match = []
        print(f"Validation Value for {column}")
        for data in data_value:
            if data not in value_consistent[column]:
                miss_match.append(data)
        print(miss_match)
```

```python
validation_churn(bank_churners)
```

```python
import re
def extract_income_category(row):
    # Menggunakan regular expression untuk mengekstrak angka
    match = re.findall(r'\d+', str(row['Income_Category']))
    
    # Jika ditemukan dua angka, pertama adalah min dan kedua adalah max
    if len(match) == 2:
        return int(match[0]), int(match[1])
    
    # Jika hanya satu angka, pertama adalah None dan kedua adalah max
    elif len(match) == 1:
        return 0, int(match[0])
    
    # Jika tidak ada angka, min dan max dianggap None
    else:
        return None, None
```

```python
import re
def handling_data_churn(data_churn):
    # buat kolom yang belum ada
    data_churn[['Income_Category_Min','Income_Category_Max']] = data_churn.apply(extract_income_category, axis=1, result_type = 'expand')

    # hapus kolom yang tidak diperlukan
    data_churn = data_churn.drop('Income_Category', axis =1)

    # hapus missing value
    data_churn = data_churn.dropna()

    # hapus duplikat
    data_churn = data_churn.drop_duplicates(keep='first')

    # casting data type
    data_churn['CLIENTNUM'] = data_churn['CLIENTNUM'].astype('int')
    data_churn['Income_Category_Min'] = data_churn['Income_Category_Min'].astype('int')
    data_churn['Income_Category_Max'] = data_churn['Income_Category_Max'].astype('int')


    # handle inconsitent data
    data_map = {
        'Existing_Customer': 'Existing Customer',
        'Existing_Customer___':'Existing Customer',
        'Attrited_Customer' : 'Attrited Customer',
        'Male': 'M',
        'Female': 'F',
        'High-School': 'High School'
    }

    # column Attrition_Flag
    data_churn['Attrition_Flag'] = data_churn['Attrition_Flag'].replace(data_map)

    # Gender
    data_churn['Gender'] = data_churn['Gender'].replace(data_map)

    # Education_Level
    data_churn['Education_Level'] = data_churn['Education_Level'].replace(data_map)
    
    # Filter data Marital status != "Unknown"
    data_churn = data_churn[data_churn ['Marital_Status'] != 'Unknown']
    
    # casting data type
    for column in data_churn.columns:
        if (data_churn[column].dtype == 'int64'):
            data_churn[column] = data_churn[column].astype('int')
    return data_churn
```

```python
transform_data_churn = handling_data_churn(bank_churners)
```
